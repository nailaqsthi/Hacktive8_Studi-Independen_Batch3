## Modelling

```python
#building K-Means Model
kmeans=KMeans().fit(df)
```

```python
#Fitting K-Means Model
score=[]
K=range(1,25)

for i in K:
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=1)
    kmeans.fit(df)
    score.append(kmeans.inertia_)
```

Menentukan nilai K yang efektif

```python
#visualize;

plt.plot(K,score,color="red")
plt.xlabel("k value")
plt.ylabel("wcss value");
```

```python
#PCA
pca = PCA(n_components = 2) 
# Create a dataframe with the two components
df_principal = pca.fit_transform(df) 
df_principal = pd.DataFrame(df_principal) 
df_principal.columns = ['PCA1', 'PCA2'] 
df_principal.head(10)
```

```python
n_clusters=20
WCSS=[]
for i in range(1,n_clusters):
    kmean= KMeans(i)
    kmean.fit(df_principal)
    WCSS.append(kmean.inertia_)    
plt.plot(WCSS, 'g*-')
plt.title('Finding the right number of clusters')
plt.xlabel('Cluster')
plt.ylabel('WCSS') 
plt.show()
```

```python
#menentukan nilai silhoutee
silhouette_scores = [] 
for n_cluster in range(2, 8):
    silhouette_scores.append(   
        silhouette_score(df_principal, KMeans(n_clusters = n_cluster).fit_predict(df_principal))) 
    
k = [2, 3, 4, 5, 6,7] 
plt.plot(k, silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 10) 
plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()
```

Silhoutte score digunakan untuk menentukan jumlah cluster yang efektif dan sesuai, berdasarkan perhitungan silhoutte score tersebut didapat nilai cluster nya adalah 2

```python
db_default = KMeans(n_clusters=2, init='k-means++').fit(df_principal) 
labels = db_default.labels_ 
```

```python
#Fit cluster into data
clusters=pd.concat([df, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head(10)
```

Berdasarkan proses K-Means Clustering yang telah dilakukan, diperoleh nilai Cluster yang efektif untuk 9000 pengguna kartu kredit adalah 2 cluster. Jumlah Cluster tersebut didasarkan pada perhitungan silhoutte score. 

Jadi setiap nasabah dikelompokkan berdasarkan Cluster 0 dan Cluster 1
